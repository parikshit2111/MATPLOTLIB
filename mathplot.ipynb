# Options for Subplot

```python
#subplot1
fig,((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5));
ax1.plot(x,x/2);
ax2.plot(np.sin(x))
ax3.scatter(np.random.random(10),np.random.random(10))
ax4.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax1.scatter(np.random.random(10),np.random.random(10))
```

```python
#subplot2 plotting on same graphs
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5));
#plot to each different axis
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[0,1].plot(x,x/2);
```

## OO METHOD

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);

# ax.set_xlim([45,100]);
```

```python
fig,ax=plt.subplots(figsize=(10,6));

scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])

# customize
ax.set(title="heart and colestrol level",
xlabel="age",ylabel="cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

#add horizontalline
ax.axhline(over_50["chol"].mean(),linestyle='dashed');
```

```python
#subplot
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))


#adddata
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#customizing

ax0.set(title="hetart disease",xlabel="age",ylabel="chol");
#legend
ax0.legend(*scatter.legend_elements(),title="Target")
#axline
ax0.axhline(over_50["chol"].mean(),linestyle='--');


#for ax1

scatter=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
#customizing

ax1.set(title="hetart disease and rate",xlabel="age",ylabel="heartrate");
#legend
ax1.legend(*scatter.legend_elements(),title="Target")
#axline
ax1.axhline(over_50["thalach"].mean(),linestyle='--');

#add title to entire figure
fig.suptitle("heart disease",fontsize=16,fontweight="bold");
```
